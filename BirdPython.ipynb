```python
from matplotlib import pyplot as plt
species = bird["Species"]
Adjusted_Trend = bird["AdjustedTrend"]
fig, ax = plt.subplots(figsize =(16, 16))
ax.barh(species, Adjusted_Trend)
for s in ['top', 'bottom', 'left', 'right']:
    ax.spines[s].set_visible(False)

ax.xaxis.set_tick_params(pad = 5)
ax.yaxis.set_tick_params(pad = 10)
 
# Add x, y gridlines
ax.grid(b = True, color ='grey',
        linestyle ='-.', linewidth = 0.5,
        alpha = 0.2)
 
# Show top values
ax.yaxis_inverted()

# Add annotation to bars
for i in ax.patches:
    plt.text(i.get_width()+0.2, i.get_y()+0.5,
             str(round((i.get_width()), 2)),
             fontsize = 10, fontweight ='bold',
             color ='grey')

ax.set_xlabel("Adjusted Trend")

plt.tick_params(axis='y', which='major', labelsize=9)
plt.tight_layout()
plt.legend()
plt.show()
```

```python
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
import matplotlib.font_manager as fm
fig, ax = plt.subplots(figsize =(20, 20))
species1 = bird[bird["Habitat"].str.contains("grassland")]
ax.barh(species1["Species"], species1["AdjustedTrend"], label = 'Grassland')
species2 = bird[bird["Habitat"].str.contains("wetland")]
ax.barh(species2["Species"], species2["AdjustedTrend"], label = "Wetland")
species3 = bird[bird["Habitat"].str.contains("urban")]
ax.barh(species3["Species"], species3["AdjustedTrend"], label = "Urban")
species4 = bird[bird["Habitat"].str.contains("successional/scrub")]
ax.barh(species4["Species"], species4["AdjustedTrend"], label = "Successional/Scrub")
species5 = bird[bird["Habitat"].str.contains("woodland")]
ax.barh(species5["Species"], species5["AdjustedTrend"], label = "Woodland")
for bars in ax.containers:
    ax.bar_label(bars)
ax.grid(b = True, color ='black',
        linestyle ='-.', linewidth = 0.5,
        alpha = 0.2)
font = fm.FontProperties(family='Times New Roman',
                                   weight='bold',
                                   size=15)

leg = ax.legend(labelspacing = 2, borderpad = 1, prop = font, loc = 'upper right')
leg.set_title("Habitat", prop = {'size': 20, 'family': "Times New Roman", 'weight':"bold"})
ax.set_xlabel("Adjusted Trend", fontsize = 20, fontweight="bold", fontname = 'Times New Roman')
ax.set_title("Trends of Maryland Bird Species", fontsize = 30, fontweight="bold", fontname = 'Times New Roman')

```

```python
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
import matplotlib.font_manager as fm
fig, ax = plt.subplots(figsize =(20, 20))
species1 = bird[bird["Habitat"].str.contains("grassland")]
ax.barh(species1["Species"], species1["AdjustedTrend"], label = 'Grassland')
species2 = bird[bird["Habitat"].str.contains("wetland")]
ax.barh(species2["Species"], species2["AdjustedTrend"], label = "Wetland")
species3 = bird[bird["Habitat"].str.contains("urban")]
ax.barh(species3["Species"], species3["AdjustedTrend"], label = "Urban")
species4 = bird[bird["Habitat"].str.contains("successional/scrub")]
ax.barh(species4["Species"], species4["AdjustedTrend"], label = "Successional/Scrub")
species5 = bird[bird["Habitat"].str.contains("woodland")]
ax.barh(species5["Species"], species5["AdjustedTrend"], label = "Woodland")
trend1 = bird[bird["OverallTrend"].str.contains("nonsignificant")]
ax.barh(trend1["Species"], trend1["AdjustedTrend"], label = 'Nonsignificant Trend', hatch=r"\\", fill = False)
for bars in ax.containers:
    ax.bar_label(bars)
ax.grid(b = True, color ='black',
        linestyle ='-.', linewidth = 0.5,
        alpha = 0.2)
font = fm.FontProperties(family='Times New Roman',
                                   weight='bold',
                                   size=15)
leg = ax.legend(labelspacing = 2, borderpad = 1, prop = font, loc = 'upper right')
leg.set_title("Habitat", prop = {'size': 20, 'family': "Times New Roman", 'weight':"bold"})
ax.set_xlabel("Adjusted Trend", fontsize = 20, fontweight="bold", fontname = 'Times New Roman')
ax.set_title("Trends of Maryland Bird Species", fontsize = 30, fontweight="bold", fontname = 'Times New Roman')
```
